### 1st insight:
    Firstly, I want to check the possibility of the people, who is not well-educated, can earn more than 50k$/month. In each major group of education, the percentage of getting more than 50k$ and less than or equal 50k$

    Reason: in order to know the impact of education level on income, higher edcuation levels get higher paid than lower ones?

```python
#Finding out the major educational groups:
#As the result, HS-grad, Some-college and Bachelors are major groups
(df['education'].value_counts()/df['education'].count())*100
```

```python
value_counts_1 = df[['education', 'income']].value_counts()
value_counts_df_1 = pd.DataFrame(value_counts_1)
value_counts_df_1 = value_counts_df_1.reset_index()
value_counts_df_1.columns = ['edu', 'income', 'counts'] # change column names
value_counts_df_1
```

```python
res_1 = pd.DataFrame(columns=['edu','income','counts'])
res_1 = value_counts_df_1[value_counts_df_1.edu == 'Bachelors']
res_1 = pd.concat([res_1 ,value_counts_df_1[value_counts_df_1.edu == 'HS-grad']],ignore_index=True)
res_1 = pd.concat([res_1, value_counts_df_1[value_counts_df_1.edu == 'Some-college']], ignore_index=True)
```

```python
res_1 = res_1.set_index('edu').groupby('edu').apply(lambda x: x.to_dict('r')).to_dict()
```

```python
#convert to json object with key as question and value as dict
key_1 , val_1 = ('Income percentage of major educational groups', res_1)
data_1 = {}
data_1[key_1] = val_1
json_data_1 = json.dumps(data_1)
json_data_1
```

```python
with open("insight_1.json", "w") as outfile:
    outfile.write(json_data_1)
```

### 2nd Insight
